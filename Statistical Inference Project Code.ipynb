```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Wholesale customers data.csv")
df.head()
```

# Fresh Visualization

```python
fresh_data = df['Fresh']
plt.figure(figsize = (10, 6))
plt.hist(fresh_data, bins = 8, color = "skyblue", edgecolor = 'black')
plt.xlabel('Annual Spending on Fresh Products (m.u.)')
plt.ylabel('Frequency')
plt.title('Histogram of Annual Spending on Fresh Products')
plt.show()
```

# Milk Visualization

```python
milk_data = df['Milk']
plt.figure(figsize = (10, 6))
plt.hist(milk_data, bins = 10, color = "skyblue", edgecolor = 'black')
plt.xlabel('Annual Spending on Milk Products (m.u.)')
plt.ylabel('Frequency')
plt.title('Histogram of Annual Spending on Milk Products')
plt.show()
```

# Grocery Visualization

```python
grocery_data = df['Grocery']
plt.figure(figsize = (10, 6))
plt.hist(grocery_data, bins = 8, color = "skyblue", edgecolor = 'black')
plt.xlabel('Annual Spending on Grocery Products (m.u.)')
plt.ylabel('Frequency')
plt.title('Histogram of Annual Spending on Grocery Products')
plt.show()
```

# Frozen Visualization

```python
frozen_data = df['Frozen']
plt.figure(figsize = (10, 6))
plt.hist(frozen_data, bins = 8, color = "skyblue", edgecolor = 'black')
plt.xlabel('Annual Spending on Frozen Products (m.u.)')
plt.ylabel('Frequency')
plt.title('Histogram of Annual Spending on Frozen Products')
plt.show()
```

# Channel Visualization

```python
horeca_data = df[df['Channel'] == 1]
horeca_count = len(horeca_data)

retial_data = df[df['Channel'] == 2]
retail_count = len(retial_data)

channels = ['Horeca', 'Retail']
channel_counts = [horeca_count, retail_count]

plt.figure(figsize = (10, 6))
plt.bar(channels, channel_counts, color = ['skyblue', 'orange'])
plt.xlabel('Channel')
plt.ylabel('Count')
plt.title('Number of Occurrences by Channel')
plt.show()
```

# Gender Visualization

```python
male_data = df[df['gender'] == 0]
male_count = len(male_data)

female_data = df[df['gender'] == 1]
female_count = len(female_data)

genders = ['Male', 'Female']
gender_counts = [male_count, female_count]

plt.figure(figsize = (10, 6))
plt.bar(genders, gender_counts, color = ['skyblue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Number of Occurrences by Gender')
plt.show()
```

# Fresh vs Milk Annual Spending

```python
plt.figure(figsize = (10, 6))
plt.scatter(fresh_data, milk_data, alpha = 0.5)
plt.xlabel('Annual Spending on Fresh products')
plt.ylabel('Annual Spending on Milk products')
plt.title('Scatter Plot of Fresh vs Milk Annual Annual Spending')
plt.grid(True)
plt.show()
```

# Average Annual Spending on Grocery per Gender

```python
gender_vs_grocery = df.groupby('gender')['Grocery'].mean()

plt.figure(figsize = (10, 6))
gender_vs_grocery.plot(kind = 'bar', color = ['skyblue', 'orange'])
plt.title('Average Annual Spending on Grocery by Gender')
plt.xlabel('Gender (0 = Male, 1 = Female)')
plt.ylabel('Mean Annual Spending on Grocery per Gender')
plt.xticks(rotation = 0)
plt.show()
```
